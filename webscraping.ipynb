```python
root = "https://quotes.toscrape.com/page/"
quotes_ =[]
tags_ = []
authors_ = []

for pages in range(1,11):

    html = requests.get(root + str(pages))

    soup = BeautifulSoup(html.text)

    for i in soup.findAll("div",{"class":"quote"}):
        quotes_.append((i.find("span",{"class":"text"})).text)

    for j in soup.findAll("div",{"class":"quote"}):
        authors_.append((j.find("small",{"class":"author"})).text)

    for k in soup.findAll("div",{"class":"tags"}):
        tags_.append((k.find("meta"))['content'])
```

```python
df = pd.DataFrame({"Quotes":quotes_,"Authors":authors_,"Tags":tags_})
```
